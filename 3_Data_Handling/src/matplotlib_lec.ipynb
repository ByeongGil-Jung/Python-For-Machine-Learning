```python
""" Matplotlib """
# pyplot 객체를 사용하여 데이터를 시각화
# pyplot 객체에 그래프들을 쌓은 다음 show() 로 flush

# Graph 는 figure 객체에 생성됨 (pyplot -> figure)
# pyplot 객체 사용시, 기본 figure 에 그래프가 그려짐
# (즉, figure 가 그림을 출력하는 일종의 인스턴스)

# plt.show() 를 한 순간, figure 를 출력하고, 현재까지의 figure 를 초기화 시킨다.
# (즉, plt.show() 를 한 순간까지의 그래프를 출력한다.)

# << 단점 >>
# arg 를 kwargs 로 받는다.
# 즉, 고정된 arg 가 없어서 alt+tab 으로 확인이 어렵다.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X = range(100)
Y = [value**2 for value in X]
plt.plot(X, Y)
plt.show()
```

```python
# 여러 그래프를 입력받을 경우, 중첩되어 출력된다.
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
# figure 안에 또 다른 figure 를 넣어서, 여러개의 figure 를 출력할 수 있다.
# subplot(r, c, i) 의 인자는, r: 행의 수, c: 열의 수, i: 출력 할 plot 의 순서이다.
# 또한 title, color, linestyle 등을 지정할 수 있다.

fig = plt.figure()  # figure 반환
fig.set_size_inches(10, 5)  # 크기 지정

plt.title("Two Lines")  # Title 설정 가능하다.

ax_1 = fig.add_subplot(1, 2, 1)  # 첫 번째 figure 생성 후 첫번쨰(1) 에 할당
ax_2 = fig.add_subplot(1, 2, 2)  # 두 번째 figure 생성 후 두번째(2) 에 할당

ax_1.plot(X_1, Y_1, c="b", ls="dashed")  # linestyle 을 지정 가능하다.
ax_2.plot(X_2, Y_2, c="#000000", ls="dotted")  # 색은 rgb 값으로 넣어도 된다.

plt.show()
```

```python
# Latex 타입의 표현도 가능 (수식 표현 가능)
plt.title("$y = \sin x$")

plt.plot(X_1, Y_1)
plt.show()
```

```python
# legend 함수로 범례를 표심, loc 위치 등 속성 지정 가능
# legend() 는 default 로 왼쪽 상단에 표기된다. (shadow, fancybox 등 옵션 가능)
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, label="line_1")
plt.plot(X_2, Y_2, label="line_2")

plt.legend(shadow=True, fancybox=True, loc="lower right")
plt.show()
```

```python
""" Matplotlib Graph """
# << Scatter >>
# - 산포도를 그릴 수 있음
# scatter 함수 사용
# maker: 데이터의 scatter 모양 지정

data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:, 0], data_1[:, 1], c="b", marker="x")
plt.scatter(data_2[:, 0], data_2[:, 1], c="r", marker="^")

plt.show()
```

```python
# s: 데이터의 크기를 지정, 데이터의 크기 비교 가능
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # pi*r^2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)  # alpha 는 투명도다.
plt.show()
```

```python
# << Bar Chart >>
# - 막대 그래프를 그릴 수 있음
# bar 함수 사용

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

X = np.arange(4)
width = 0.25

plt.bar(X, data[0], color="b", width=width)
plt.bar(X + width, data[1], color="r", width=width)
plt.bar(X + width*2, data[2], color="g", width=width)
plt.xticks(X + width, ("A", "B", "C", "D"))  # 이름 설정

plt.show()
```

```python
# 위로 그래프를 쌓을 경우

data = np.array([[5., 25., 50., 20.],
                 [4., 23., 51., 17.],
                 [6., 22., 52., 19.]])
# print(data.shape[0]) // 3
# print(data.shape[1]) // 4

color_list = ["b", "g", "r"]
data_label = ["A", "B", "C"]
X = np.arange(data.shape[1])

for i in range(data.shape[0]):
    plt.bar(X, data[i], bottom=np.sum(data[:i], axis=0), color=color_list[i], label=data_label[i])
    # 왜 np.sum(data[:i]) 이지 ?? -> y 의 좌표이기 때문, axis=0 은 같은 col 끼리의 sum 이다.
    # 즉, 누적된 col 의 값을 축적해서 나타낸다. (이렇게 쌓아 올려간다.)

plt.legend()
plt.show()
```

```python
# << Histogram >>
# - 히스토그램을 그릴 수 있음
# hist 함수 사용
# bins : 데이터를 나누는 갯수이다.

X = np.random.randn(1000)  # randn 은 normal dist 를 따르는 random 변수이다.
plt.hist(X, bins=100)
plt.show()
```

```python
# << Boxplot >>
# - 박스플롯을 그릴 수 있음
# boxplot 함수 사용
# 50% 구간을 네모로 표시하며, 내부의 선은 Median(중간값) 이다.
# range 에서 벗어난 값들은 점으로 표시한다.

data = np.random.randn(100, 5)  # 뒤의 5는 5개의 row 를 만든다.
plt.boxplot(data)
plt.show()
```
